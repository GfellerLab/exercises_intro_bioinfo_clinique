```python
# Simple analysis with python
import pandas as pd                                                                                                                                                                                       
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_context("paper")
                                                                                                                                                                                                          
df = pd.read_csv("./04aa322d-117d-4b83-8b26-ad0b04299275.wxs.aliquot_ensemble_masked.maf", sep = '\t', comment = '#')  
#let's have a look at the dataset
print(df)
# The files contains many columns(140).. try to understand which are the most important
```

```python
#Q1) which chromosome are the most mutated?                                                                                                                                                                
print(df['Chromosome'].value_counts()) 
fig, ax = plt.subplots(figsize=(12, 6))
sns.countplot(data = df, x = 'Chromosome', ax = ax)   
```

```python
# Many different types of mutations
fig, ax = plt.subplots(figsize=(12, 6))
print(df['Variant_Classification'].value_counts())
# Q2) explain the next two plots
sns.countplot(data = df, x = 'Variant_Classification', ax = ax) 
fig, ax = plt.subplots(figsize=(12, 6))                                                                                                  
sns.countplot(data = df, x = 'Chromosome', hue = 'Variant_Classification', ax = ax)
```
