## Données de séquence pour étudier les mutations cancéreuses

Pendant le cours, vous avez étudié comment, dans le séquençage à haut débit, les reads doivent être mappés sur le génome humain.
On va faire un calcul simple pour estimer le nombre d'opérations qui sont nécessaires. 

Le génome humain contient environ 20 000 gènes codants. Supposons que la longueur moyenne des exons soit de 10000 nucléotides.

**Question 1)** Quel est le nombre approximatif d'opérations pour mapper un read ? Et 10^6 reads (le output typique d'une machine à haut débit) ? 

Vous n'avez pas vraiment besoin de connaître la longueur d'un read pour résoudre l'exercice, mais, pour votre connaissance, la longueur d'un read dépend de la technologie de séquençage. La longueur typique pour le séquençage Illumina (l'une des plus utilisées dans le monde) est de ~80 paires de bases [80bp].

```python
################################################
#Solution Question 1)
################################################
# pour un read -> 20000 (num gènes codants) * 10000 (taille exon) = 2 10^7 base pairs ~= 2^24 base pairs
# Le nombre d'opérations pour mapper naïvement un read sur le génome est : 
# N_op = longueur_génome - longueur_read + 1 ~= longueur_génome ~= 2^24 opérations

# pour 10^6 reads, vous devez répéter cette recherche pour chaque read donc
# N_op = 10^6 * 2^24 ~= 2^43. A huge number! 
```

Voici à quoi ressemble la sortie d'une machine de séquençage (file *.fastq*). 

**Question 2)** Pouvez-vous dire la signification de chaque ligne ?

```python
################################################
#Solution Question 2)
################################################
# La ligne 1 commence par un caractère '@' et est suivie d'un identificateur de séquence et d'une description facultative.
# La ligne 2 correspond aux lettres de la séquence brute.
# La ligne 3 commence par un caractère "+" et est éventuellement suivie du même identificateur de séquence (et d'une description éventuelle).
# La ligne 4 code les valeurs de qualité pour la séquence de la ligne 2, et doit contenir le même nombre de symboles que de lettres dans la séquence.
```
